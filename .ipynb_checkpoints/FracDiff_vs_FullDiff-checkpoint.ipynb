# Fractional differentiation

Fractional differentiation is a mathematical operation where a function is differentiated to a fractional order, rather than an integer order as is commonly done in calculus. In the context of financial returns, fractional differentiation can be used to transform time series data in a way that retains memory of past information, which can be crucial for modeling and forecasting.

In financial time series analysis, the goal is often to make the series stationary in the mean (constant mean over time) so that various statistical and machine learning models can be applied. Traditional full differentiation of order 1 (i.e., differencing) can be used to achieve stationarity, but it often removes too much memory (correlation between consecutive variables) from the series, which can be detrimental for predictive modeling.

Full differentiation, on the other hand, refers to the process of differentiating a function to an integer order. In the context of financial returns, this would typically involve calculating the change in price from one time period to the next, which can help to identify trends and patterns in the data. However, as mentioned earlier, this process can remove a lot of the historical information contained in the series.

Let's see if we can find a tradeoff between non-stationarity and losing too much memory of the time series by computing the first differences.

## Comparison
**Memory Preservation:**

- *Fractional Differentiation*: Preserves more memory of the series, which can be beneficial for modeling and forecasting.
- *Full Differentiation*: Tends to remove a significant amount of memory from the series.

**Stationarity:**

- *Fractional Differentiation*: Allows for a more controlled transformation to stationarity, potentially retaining more useful information.
- *Full Differentiation*: Often achieves stationarity but at the cost of losing potentially useful information.

**Complexity:**

- *Fractional Differentiation*: More complex to implement as it involves fractional orders of differentiation.
- *Full Differentiation*: Simpler to implement as it involves integer orders of differentiation.

**Application in Financial Analysis:**

- *Fractional Differentiation*: Can be used to develop more sophisticated models that take into account the memory-preserving properties of fractional differentiation.
- *Full Differentiation*: Typically used in simpler models where the focus is on identifying trends and patterns in the data.

In essence, if you retain more information that would be better when searching for patterns among variables.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from twdm import tqdm

import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
df = pd.read_csv("data/Rv_daily_lec4.csv", index_col=0)
```

The function `def getWeights_FFD` calculates the weights that are used to compute the fractional differences of a time series.

It takes two parameters: `d`, which is the fractional differentiation order, and `thr`, which is a threshold value to determine the minimum weight to consider.

The function starts with an initial weight of 1 and iteratively calculates new weights using the formula `-w[-1]/k*(d-k+1)` until the absolute value of the newly calculated weight is less than the threshold `thr`.

The weights are then reversed and reshaped into a column vector before being returned.

You can have a look at the reading provided in the syllabus to revise the theory behind these calculations.

```python
def getWeights_FFD(d, thr=1e-5) -> np.array:
    """ Calculates the weights to convolute a time series with to calculate the numerical fractional differences
    of a certain length determined by when the weights drop below the threshold, thr."""
    assert d>0, "fractional dimension, d, must be postive!"
    # w>thr drops insignificant weights
    w=[1.]
    k=1
    while np.abs(w[-1])>thr:
        w_=-w[-1]/k*(d-k+1)
        k=k+1
        w.append(w_)
        
    w=np.array(w[::-1]).reshape(-1,1)
    return w
```

The function `fracDiff_FFD` calculates the fractional differences of each column in a given DataFrame.

It takes three parameters: 
- `input_dfx`, which is a DataFrame containing the time series data
- `d`, which is the fractional differentiation order
- `thres` which is a threshold value for determining the cut-off weight for the longest window.

The function first calls `getWeights_FFD` to get the weights for the fractional differentiation. It then iterates over each column in the input DataFrame, forward fills NaN values, and computes the fractional differences by applying the weights to a moving window of values in the series. The computed fractional differences are then stored in a new DataFrame which is returned at the end.

```python
def fracDiff_FFD(input_df: pd.DataFrame, d, thres=1e-5):
    '''
    Calculates numerical fractional differences of each column in the dataframe or series.
    Note 1: thres determines the cut-off weight for the longest window
    Note 2: d can be any positive fractional, not necessarily bounded [0,1].
    '''
    assert d>0, "fractional dimension, d, must be postive!"
    #1) Compute weights for the longest series
    w=getWeights_FFD(d, thres)
    width=len(w)-1
    #2) Apply weights to values
    df={}
    for name in input_df.columns:
        seriesF = input_df[[name]].fillna(method='ffill').dropna()
        df_ = pd.Series(dtype=float)
        for iloc1 in range(width, seriesF.shape[0]):
            loc0, loc1 = seriesF.index[iloc1-width],seriesF.index[iloc1]
            if not np.isfinite(input_df.loc[loc1, name]):continue  # skip NAs
            df_[loc1] = np.dot(w.T, seriesF.loc[loc0:loc1])[0,0]
        df[name] = df_.copy(deep=True)
        
    df=pd.concat(df,axis=1)
    return df
```

The following code snippet visualizes the process of fractionally differentiating a time series (logarithm of SP500 closing prices in this case) until it becomes stationary. The stationarity is tested using the Augmented Dickey-Fuller (ADF) test and the code stop using fractional difference as soon as the ADF test evaluate the transformed series as stationary. Note that we use the log of prices to be able to visualize all the time series in the same plot.

```python
fig = plt.figure(figsize=(15, 10))
axes = fig.subplots()

not_stationary = True
sp500 = df[['SP_close']]
logsp500 = np.log(sp500)
logsp500.columns = [f'log({sp500.columns[0]})']
d = 0

logsp500.plot(secondary_y=False, logy=False, ax=axes)  # important to use log prices by the way!

while not_stationary:
    d = d + 0.1
    fd_sp = fracDiff_FFD(pd.DataFrame(logsp500), d, thres=1e-5)
    result = adfuller(fd_sp)
    print(f"fractionally differentiated  at d={d:0.2f}, ADF p-value={result[1]:.3E}")
    fd_sp.columns = [f'{fd_sp.columns[0]} differentiated d={d:.1f}, ADF p-value={result[1]*100:.1f}%']
    ax1 = fd_sp.plot(secondary_y=True, logy=False, alpha=0.2, ax=axes)

    not_stationary = result[1]>0.05 # p-value
axes.legend(loc='lower left')
ax1.legend(loc='upper left')
```

```python
# Loop to fractionally differentiate series from 0.1 to 0.9
toconcat = []
d_values = np.arange(0.1, 1.0, 0.1)
for d in tqdm(d_values):
    fd_sp = fracDiff_FFD(df[['SP_close']], d, thres=1e-5)
    fd_sp.columns = [f'frac_diff_{d:.1f}']
    toconcat.append(fd_sp)

# Add fully differentiated series (d=1)
ret = df[['SP_close']].pct_change(1)
ret.columns = ['full_diff']
toconcat.append(ret)
toconcat.append(df[['SP_close']])

fd_df = pd.concat(toconcat,axis=1).sort_index()
fd_df.head()
```

```python
fd_df.sort_index().isnull().sum()
```

The reason you are observing fewer non-null values as you decrease the value of `d` in the fractional differentiation process is due to the increasing window size required to compute the fractional differences.

In the `fracDiff_FFD` function, the weights for the fractional differentiation are computed using the `getWeights_FFD` function. The number of weights (or the window size) increases as `d` decreases. This is because the weights decay more slowly, and it takes a larger number of terms to reach the threshold value (`thres=1e-5` in your case) which determines when to stop adding more weights.

Therefore, when you compute the fractional differences with a smaller `d`, a larger window of data points is required to compute each value in the series, resulting in a larger number of `NaN` values at the beginning of the series. This is why you see fewer non-null values in the output DataFrame for smaller values of `d`.

Notice below that the correlation between the fractionally differentiated series and the initial close price series decreases as `d` increases.

```python
# Calculate the correlation matrix
corr_matrix = fd_df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
