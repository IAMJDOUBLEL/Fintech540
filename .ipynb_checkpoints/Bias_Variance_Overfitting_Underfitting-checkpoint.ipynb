# Bias-Variance and Underfitting-Overfitting connection

In the context of machine learning, the concepts of **bias-variance tradeoff** and **overfitting/underfitting** are closely related. Here's a concise explanation:

**Bias-Variance Tradeoff**

*Bias*: This refers to the error due to overly simplistic assumptions in the learning algorithm, which can lead it to underfit the data. High bias can cause the model to miss relevant relations between features and target outputs (it is not flexible enough to learn from the data). This way the model will be on average far from predicting the actual value.

*Variance*: This refers to the error due to too much complexity in the learning algorithm, which can lead it to overfit the data. High variance can cause the model to model the random noise in the data, which means it is too flexible and adapts too much to the data.

The tradeoff exists because if you want to lower the bias you have to make the model more complex. If you make the model too complex you are going to have high variance.

**Overfitting/Underfitting**

*Overfitting*: This occurs when the model learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on new data. This is generally a result of *high variance and low bias*.

*Underfitting*: This occurs when the model cannot capture the underlying trend of the data. This is generally a result of *high bias and low variance*.

**Connection**

The connection between the **bias-variance** tradeoff and **overfitting/underfitting** is that when you are trying to minimize the bias (*making the model more complex to fit the data closely*), you may increase the variance, which can lead to overfitting. Conversely, when you try to minimize the variance (*making the model simpler*), you may increase the bias, which can lead to underfitting. Therefore, achieving a good balance between bias and variance is crucial to developing models that generalize well to new, unseen data.

The follow is an exercise that visually demonstrates the bias-variance tradeoff and the overfitting/underfitting concepts using a polynomial regression example:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

Let's generate some synthetic data using a polynomial equation with some noise added to y. In this case you can see that y depends on x by construction. The `.reshape(-1,1)` method  is taking care of reshaping the arrays to have a single column (necessary for the train_test_split function).

The equation that relates $x$ to $y$ in the given code can be written as follows:

$$
y = x - 0.04x^2 + 0.03x^3 + 0.05x^5 + \epsilon
$$

where:
- $y$ is the dependent variable (output).
- $x$ is the independent variable (input).
- $\epsilon$ is a noise term added to $y$, which follows a normal distribution with a mean of 0 and a standard deviation equal to 1% of the standard deviation of $y$ (before adding noise).

The noise term $\epsilon$ is generated as follows:

$$
\epsilon \sim N(0, \sigma)
$$

where:
- $\mathcal{N}(0, \sigma)$ denotes a normal distribution with mean 0 and a standard deviation $\sigma$ calculated as 1% of the standard deviation of $Y$ (before adding noise), which is represented by the variable `noise_scale` in the code.



```python
np.random.seed(0)
n = 200 # number of data points to generate
x = 2 - 3 * np.random.normal(0, 1, n)
y = x - 0.04 * (x ** 2) + 0.03 * (x ** 3) + 0.05 * (x ** 5)
noise_scale = np.std(y) * 0.01  
noise = np.random.normal(0, noise_scale, len(y))
y += noise

x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

Then splits the data into training and testing sets, with 80% of the data used for training and 20% for testing. The `random_state` parameter ensures that the split is reproducible.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

Here, a list of polynomial degrees is defined to demonstrate the effects of underfitting and overfitting. A figure is initialized with a specified size, and a loop starts to create subplots for each degree.

Within the loop, for each degree, polynomial features are generated for both the training and testing sets using the `PolynomialFeatures` class. If you inspect `x_poly_train` or `x_poly_test`, you observe that the higher the degree, the more features we are creating to fit the initial data. When the degree is 1, data stays the same (`x_train == x_poly_train`), except for the fact that we are also adding a constant, so you can at most fit a line.

Then, a linear regression model is initialized and fitted to the polynomial features of the training set. Predictions are then made on the train and test set. This two-step procedure is the same as fitting the polynomial regression directly, but should give a better idea of what we are doing here.

The train and test set data points are plotted as a pair $(x,y)$ on a scatter plot together with the actual values. The title of each subplot displays the degree of the polynomial and the mean squared error (MSE) calculated from the predictions and the true values of the train\test set. Please note that in-sample and out-of-sample is equivalent to say train and test.

```python
degrees = [1, 2, 4, 10, 12]
```

```python
np.random.seed(100)
fig, axes = plt.subplots(2, len(degrees), figsize=(14, 10))

for i in range(len(degrees)):
    # In-sample testing (Training data)
    ax = axes[0, i]
    polynomial_features = PolynomialFeatures(degree=degrees[i])
    x_poly_train = polynomial_features.fit_transform(x_train)
    x_poly_test = polynomial_features.transform(x_test)

    model = LinearRegression()
    model.fit(x_poly_train, y_train)
    y_poly_pred_train = model.predict(x_poly_train)

    ax.scatter(x_train, y_train, s=20, alpha=0.5, label='true')
    ax.scatter(x_train, y_poly_pred_train, s=20, label='fitted')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc=0)
    ax.set_title("In-sample\nDegree {}\nMSE = {:.2e}".format(degrees[i], np.mean((y_poly_pred_train - y_train)**2)))

    # Out-of-sample testing (Testing data)
    ax = axes[1, i]
    y_poly_pred_test = model.predict(x_poly_test)

    ax.scatter(x_test, y_test, s=20, alpha=0.5, label='true')
    ax.scatter(x_test, y_poly_pred_test, s=20, label='fitted')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc=0)
    ax.set_xlim(-6,11)
    ax.set_ylim(-500,4000)
    ax.set_title("Out-of-sample\nDegree {}\nMSE = {:.2e}".format(degrees[i], np.mean((y_poly_pred_test - y_test)**2)))
    
fig.tight_layout()
```

A linear regression model is being fitted to the data, but the data itself has been transformed to include polynomial features. This is why we can approximate non-linear relationships well despite using a linear regression model. 

As you can see, the higher the polynomial degree, the better you fit the synthetic data: the MSE calculated between the predicted and the actual value goes down with the degree. A visual inspection also helps us understand that we are achieving a better fit if we use a polynomial regression of degree 15 rather than degree 1. Beyond a certain degree, we don't see any visual difference in the fitting by using the human eye, but there is a difference in the fitting that the MSE quantifies. **The takeaway is that we are reducing the bias by increasing the degree of the polynomial**. The model fitted also performs well out-of-sample since the type of noise stays the same in the test set (remember that we created $y$ from $x$ and then split that $y$ into two different sets. This is not a common situation because, in general, you do not deal with synthetic data).

What happens to the out-of-sample fitting if you add additional noise to the data? This is equivalent to a financial scenario with a regime change, and the data starts changing distribution (additional noise is created due to an external event). Adding noise to $y$ simulates the variability you might see in real-world observations, where many unobserved factors can influence the response variable; hence it is often perturbed with noise to simulate this uncertainty. This can help visualize how the model performs under more realistic conditions where the data might be variable.

```python
np.random.seed(100)
fig, axes = plt.subplots(2, len(degrees), figsize=(16, 12))

for i in range(len(degrees)):
    # In-sample testing (Training data)
    ax = axes[0, i]
    polynomial_features = PolynomialFeatures(degree=degrees[i])
    x_poly_train = polynomial_features.fit_transform(x_train)
    x_poly_test = polynomial_features.transform(x_test)

    model = LinearRegression()
    model.fit(x_poly_train, y_train)
    y_poly_pred_train = model.predict(x_poly_train)

    ax.scatter(x_train, y_train, s=20, alpha=0.5, label='true')
    ax.scatter(x_train, y_poly_pred_train, s=20, label='fitted')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc=0)
    ax.set_title("In-sample\nDegree {}\nMSE = {:.2e}".format(degrees[i], np.mean((y_poly_pred_train - y_train)**2)))

    # Out-of-sample testing (Testing data)
    ax = axes[1, i]
    noise_scale = np.std(y_train) * 0.08  # Setting noise scale as 8% of the standard deviation of y_train
    noise = np.random.normal(0, noise_scale, len(x_test))  
    y_test_noisy = y_test + noise[:, np.newaxis]
    y_poly_pred_test = model.predict(x_poly_test)

    ax.scatter(x_test, y_test_noisy, s=20, alpha=0.5, label='true (with noise)')
    ax.scatter(x_test, y_poly_pred_test, s=20, label='fitted')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(loc=0)
    ax.set_xlim(-6,11)
    ax.set_ylim(-500,4000)
    ax.set_title("Out-of-sample\nDegree {}\nMSE = {:.2e}".format(degrees[i], np.mean((y_poly_pred_test - y_test_noisy)**2)))

fig.tight_layout()
```

Suppose you compare the out-of-sample fitting by looking at the MSE before and after perturbing the `y_test` with additional noise. In that case, you can observe how the gain in MSE reduction is smaller when you fit a higher degree polynomial regression. This is happening because a polynomial regression of degree 12 decreases the bias a lot and hence is subjected to higher variance when the test set changes. **The takeaway is that a polynomial regression with a too high degree may overfit your data and perform poorly out-of-sample because it has high variance**. In this example, the polynomial regression of degrees 12 overfitted the initial data. When you modify the target $y$ with some noise, it cannot perform as before. On the contrary, a degree of 10 seems to maintain a reasonable level of MSE compared to the other lower degrees tested in this notebook.


On the other hand, one can notice that a linear regression (degree equal to 1) is underfitting the data and performing worse than all the other polynomial regression, even before the additional noisy perturbation. This is because the bias is high, but you can observe that the vriance is lower because the performance is similar before and after the perturbation. So we may need to adjust the performance by trading-off some low variance with the bias.
