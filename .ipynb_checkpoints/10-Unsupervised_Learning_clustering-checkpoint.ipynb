# Unsupervised Learning - Introduction to Clustering and Dimensionality Reduction

In the unsupervised learning setting, the training data consists of a set of input vectors $x$ without any corresponding target values.

Those algorithms learn patterns from data without explicit labels. Among the categories of unsupervised learning techniques we will approach two of them in this lecture.

In the unsupervised learning setting, the training data consists of a set of input vectors $x$ without any corresponding target values.

The goal in such problems may be to discover groups of similar examples within the data, which is called **clustering**, to determine the distribution of data within the input space, known as density estimation, or to project the data from a high-dimensional space down to two or three dimensions for the purpose of **dimensionality reduction**.

In this lecture, we will focus on some examples of clustering and dimensionality reduction models in machine learning, which can be enormously useful for solving problems in the financial space.

## Clustering
Clustering is the process of grouping similar data points based on their features, such that data points in the same group are more similar to each other than to those in other groups. It is a method of exploratory data analysis that can reveal hidden patterns within datasets.

**K-Means Clustering**:
K-Means is a popular centroid-based clustering algorithm that partitions a set of observations into K clusters. Each observation belongs to the cluster with the nearest mean. The process involves:
   - Initializing K centroids randomly.
   - Assigning data points to the closest centroid.
   - Recalculating the centroid as the mean of all points in the cluster.
   - Repeating the above steps until convergence.

## Dimensionality Reduction
Dimensionality reduction techniques aim to reduce the number of features in a dataset, preserving its structure. This can be useful for data visualization, compression, and speeding up machine learning algorithms.

**Principal Component Analysis (PCA)**
PCA is a statistical procedure that transforms a number of possibly correlated variables into a smaller number of uncorrelated variables called principal components. The first principal component captures the most variance in the data, and each succeeding component captures less and less.

These methods, among others in unsupervised learning, offer powerful tools to unveil hidden structures in data, aiding in various financial and analytical tasks.

## K-Means Clustering

K-means clustering is one of the simplest and most popular unsupervised machine learning algorithms. The primary goal of this algorithm is to divide a set of data points into distinct non-overlapping subgroups (clusters) based on their similarities.

**Basic Idea**

1. **Initialization**: Randomly select `k` data points (seeds) to be the initial centroids.
2. **Assignment**: Assign each data point to the nearest centroid. This forms `k` clusters.
3. **Update**: Calculate the new centroids (mean) for each cluster.
4. **Repeat**: Repeat the assignment and update steps until the centroids no longer change, or a predefined number of iterations is reached.

The algorithm tries to minimize the variance (or inertia) within each cluster. The variance within a cluster for a given centroid is typically computed using the sum of the squared Euclidean distances between the data points and the centroid.

**Mathematical Formulation**

The objective of the k-means clustering is to minimize the total intra-cluster variance, or the squared error function:

$$ J = \sum_{i=1}^{m}\sum_{j=1}^{k} w_{ij} ||x_i - \mu_j||^2 $$

Where:
- $ m $ is the number of data points.
- $ k $ is the number of clusters.
- $ x_i $ is the $ i^{th} $ data point.
- $ \mu_j $ is the centroid of the $j^{th}$ cluster.
- $ w_{ij} $ is 1 if data point $ x_i $ belongs to cluster $ j $, otherwise 0.

The algorithm iteratively updates the centroids and the cluster assignments to minimize the objective function $ J $.

It's important to note that k-means might converge to a local optimum. Therefore, it's a common practice to run the algorithm multiple times with different initializations and choose the clustering that has the lowest cost $ J $.

**K-means vs. K-Nearest Neighbor**

Both methods have k in the name, but this is **completely incidental**. 

*K* refers to something different for each method:
- the number of clusters in k-means 
- the number of neighbors in KNN . 

Hence, they are used for completely different purposes, even though there are some connections between them.

Both methods involve computing distances in input space and assigning data points to a set of nearest **prototype points**. They differ in this respect because

- In KNN, the prototypes are training points. In k-means, the prototypes are cluster centroids, which are not restricted to be data points themselves.
- In KNN, the training points are simply memorized and held fixed. In k-means, cluster centroids are learned by updating their values during training (which also changes the assignment of data points to centroids).
- In k-means, data points are assigned to a single centroid (the nearest). In KNN, data points are assigned to the k nearest training points.

*Once k-means has been trained, new points can be assigned to the existing clusters by simply finding the nearest centroid. This procedure is identical to a KNN classifier with k=1, where the training set consists of the cluster centroids and the labels consist of the cluster ids.*

### Decision Boundaries

Let's plot the model's decision boundaries. This gives us a _Voronoi diagram_:

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=30, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=50, linewidths=50,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(8, 4))
plot_decision_boundaries(kmeans, X)
plt.show()
```

Not bad! Some of the instances near the edges were probably assigned to the wrong cluster, but overall it looks pretty good.

### K-Means Variability

In the original K-Means algorithm, the centroids are just initialized randomly, and the algorithm simply runs a single iteration to gradually improve the centroids, as we saw above.

However, one major problem with this approach is that if you run K-Means multiple times (or with different random seeds), it can converge to very different solutions, as you can see below:

```python
def plot_clusterer_comparison(clusterer1, clusterer2, X, title1=None, title2=None):
    clusterer1.fit(X)
    clusterer2.fit(X)

    plt.figure(figsize=(10, 3.2))

    plt.subplot(121)
    plot_decision_boundaries(clusterer1, X)
    if title1:
        plt.title(title1, fontsize=14)

    plt.subplot(122)
    plot_decision_boundaries(clusterer2, X, show_ylabels=False)
    if title2:
        plt.title(title2, fontsize=14)
        
    return clusterer1,clusterer2
```

```python
kmeans_rnd_init1 = KMeans(n_clusters=5, init="random", n_init=1,
                         algorithm="lloyd", random_state=42)
kmeans_rnd_init2 = KMeans(n_clusters=5, init="random", n_init=1,
                         algorithm="lloyd", random_state=21)

k1,k2 = plot_clusterer_comparison(kmeans_rnd_init1, kmeans_rnd_init2, X,
                          "Solution 1", "Solution 2 (with a different random init)")
```

### Inertia

To select the best model, we will need a way to evaluate a K-Mean model's performance. Unfortunately, clustering is an unsupervised task, so we do not have the targets. But at least we can measure the distance between each instance and its centroid. This is the idea behind the _inertia_ metric:

### K-Means++

Instead of initializing the centroids entirely randomly, it is preferable to initialize them using the following algorithm, proposed in a [2006 paper](https://goo.gl/eNUPw6) by David Arthur and Sergei Vassilvitskii:
* Take one centroid $c_1$, chosen uniformly at random from the dataset.
* Take a new center $c_i$, choosing an instance $\mathbf{x}_i$ with probability: $D(\mathbf{x}_i)^2$ / $\sum\limits_{j=1}^{m}{D(\mathbf{x}_j)}^2$ where $D(\mathbf{x}_i)$ is the distance between the instance $\mathbf{x}_i$ and the closest centroid that was already chosen. This probability distribution ensures that instances that are further away from already chosen centroids are much more likely be selected as centroids.
* Repeat the previous step until all $k$ centroids have been chosen.

The rest of the K-Means++ algorithm is just regular K-Means. With this initialization, the K-Means algorithm is much less likely to converge to a suboptimal solution, so it is possible to reduce `n_init` considerably. Most of the time, this largely compensates for the additional complexity of the initialization process.

To set the initialization to K-Means++, simply set `init="k-means++"` (this is actually the default):

```python
KMeans(init="k-means++")
```

```python
good_init = np.array([[-3, 3], [-3, 2], [-3, 1], [-1, 2], [0, 2]])
kmeans = KMeans(n_clusters=5, init=good_init, n_init=1, random_state=42)
kmeans.fit(X)
kmeans.inertia_
```

No, we cannot simply take the value of $k$ that minimizes the inertia, since it keeps getting lower as we increase $k$. Indeed, the more clusters there are, the closer each instance will be to its closest centroid, and therefore the lower the inertia will be. However, we can plot the inertia as a function of $k$ and analyze the resulting curve:

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42, n_init='auto').fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]
```

```python
plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.annotate('Elbow',
             xy=(4, inertias[3]),
             xytext=(0.55, 0.55),
             textcoords='figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.axis([1, 8.5, 0, 1300])
# save_fig("inertia_vs_k_plot")
plt.show()
```

As you can see, there is an elbow at $k=4$, which means that less clusters than that would be bad, and more clusters would not help much and might cut clusters in half. So $k=4$ is a pretty good choice. Of course in this example it is not perfect since it means that the two blobs in the lower left will be considered as just a single cluster, but it's a pretty good clustering nonetheless.

```python
plot_decision_boundaries(kmeans_per_k[4-1], X)
plt.show()
```

Another approach is to look at the _silhouette score_, which is the mean _silhouette coefficient_ over all the instances. An instance's silhouette coefficient is equal to $(b - a)/\max(a, b)$ where $a$ is the mean distance to the other instances in the same cluster (it is the _mean intra-cluster distance_), and $b$ is the _mean nearest-cluster distance_, that is the mean distance to the instances of the next closest cluster (defined as the one that minimizes $b$, excluding the instance's own cluster). The silhouette coefficient can vary between -1 and +1: a coefficient close to +1 means that the instance is well inside its own cluster and far from other clusters, while a coefficient close to 0 means that it is close to a cluster boundary, and finally a coefficient close to -1 means that the instance may have been assigned to the wrong cluster.

Let's plot the silhouette score as a function of $k$:

```python
silhouette_score(X, kmeans.labels_)
```

```python
silhouette_scores = [silhouette_score(X, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.axis([1.8, 8.5, 0.55, 0.7])
# save_fig("silhouette_score_vs_k_plot")
plt.show()
```

As you can see, this visualization is much richer than the previous one: in particular, although it confirms that $k=4$ is a very good choice, but it also underlines the fact that $k=5$ is quite good as well.

An even more informative visualization is given when you plot every instance's silhouette coefficient, sorted by the cluster they are assigned to and by the value of the coefficient. This is called a _silhouette diagram_:

```python
plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    
    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)
```

### Limits of K-Means

**Spherical Clusters**
A spherical cluster implies that data points are uniformly spread out in all directions from the center of the cluster, creating a sphere-like shape. Imagine you have a cluster where the center of the cluster is equidistant from all its boundary points in all directions. The classic k-means algorithm is typically biased towards identifying such spherical clusters because it uses the mean (centroid) to represent the center of the cluster and assigns data points to the nearest centroid.

**Ellipsoidal Clusters**
In contrast, an ellipsoidal cluster stretches more in one direction than others. Picture an elongated or squashed sphere. The shape can be thought of as resulting from taking a spherical cluster and applying different levels of stretching or squeezing along different dimensions. An ellipsoidal shape can be defined using three characteristics:

- Center: The point at the heart of the ellipsoid.
- Size: Defined by the lengths of the major and minor axes.
- Orientation: The direction in which the ellipsoid stretches.

GMMs are particularly adept at identifying ellipsoidal clusters. This is because a Gaussian distribution, which is at the core of GMM, can be visualized as an ellipsoid in the feature space. The parameters of this Gaussian (mean and covariance matrix) determine the center, size, and orientation of the ellipsoid.

Real-world data often does not form perfect spherical clusters. For example, consider data points representing people's incomes and ages. These might not form perfect circles (spheres in higher dimensions) but rather elongated shapes due to various socio-economic factors.

```python
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]
```

### Gaussian Mixture Model (GMM)

A Gaussian Mixture Model (GMM) is a probabilistic model that assumes all the data points are generated from a mixture of several Gaussian distributions with unknown parameters. In simple terms, rather than having one Gaussian distribution fit all the data, a GMM will try to find multiple Gaussian distributions that, when combined, can describe the data more effectively.

A [Gaussian mixture model](https://scikit-learn.org/stable/modules/mixture.html#gmm) can actually separate these clusters pretty well (using all 4 features: petal length & width, and sepal length & width).

**Mathematical Formulation**

Given a dataset $X$, the probability $p(x)$ that a data point $x$ was generated by this model is given by:

$$ p(x) = \sum_{i=1}^{k} \phi_i \mathcal{N}(x | \mu_i, \Sigma_i) $$

Where:
- $ \phi_i $ is the weight of the $i^{th}$ Gaussian distribution.
- $ \mathcal{N}(x | \mu_i, \Sigma_i) $ is the $i^{th}$ Gaussian distribution with mean $\mu_i $ and covariance matrix $ \Sigma_i $.
- $ k $ is the number of Gaussian distributions.

The goal is to estimate the parameters $ \phi, \mu, \Sigma $ that maximize the likelihood of the model given the data.

**Expectation-Maximization (EM) for GMM**

Estimating the parameters is done using the Expectation-Maximization (EM) algorithm, which works as follows:

1. **Initialization**: Start with an initial estimate of the parameters $ \phi, \mu, \Sigma $.

2. **Expectation step (E-step)**: Assign each data point to the Gaussian distribution that is most likely to have generated it. More specifically, compute the "responsibilities" $ \gamma(z_{ik}) $ which represent the probability that instance $ x_i $ was generated by cluster $ k $.

$$  \gamma(z_{ik}) = \frac{\phi_k \mathcal{N}(x_i | \mu_k, \Sigma_k)}{\sum_{j=1}^{k} \phi_j \mathcal{N}(x_i | \mu_j, \Sigma_j)}  $$

3. **Maximization step (M-step)**: Update the model parameters to maximize the likelihood given the assigned data points. Compute:

$$ \phi_k^{new} = \frac{1}{m}\sum_{i=1}^{m} \gamma(z_{ik}) $$

$$ \mu_k^{new} = \frac{\sum_{i=1}^{m} \gamma(z_{ik}) x_i}{\sum_{i=1}^{m} \gamma(z_{ik})} $$

$$ \Sigma_k^{new} = \frac{\sum_{i=1}^{m} \gamma(z_{ik}) (x_i - \mu_k^{new})(x_i - \mu_k^{new})^T}{\sum_{i=1}^{m} \gamma(z_{ik})} $$

4. **Convergence**: Repeat the E-step and M-step until the parameters converge.

The result is a set of Gaussian distributions that model the data, and each data point is associated with one of the Gaussian distributions.

GMMs are especially useful when the clusters in the data are not spherical or have different sizes, as they can fit ellipsoidal clusters of different sizes and orientations. However, it's important to note that GMM requires a good initialization and might converge to a local optimum, so it might be beneficial to run it multiple times with different initializations.

# Principal Component Analysis
Principal component analysis (PCA) is an unsupervised learning technique that is apt to learn a lower dimensional representation of the data, partially independent (no linear correlation between variables). It learns an orthogonal linear transformation that projects the original input data into another space. It is also considered a dimensionality reduction method that preserves as much information as possible.

Performing PCA means finding and ranking all the eigenvalues and eigenvectors of a covariance matrix. This is useful because high-dimensional data (with p features) may have nearly all their variation in a small number of dimensions k, i.e., the subspace spanned by the eigenvectors of the covariance matrix that have the k largest eigenvalues. If we project the data into this subspace, we can have a dimension reduction (from p to k) with hopefully little loss of information.

 **We tend to do this because working with less variable is better, more manageable, but we do not want to lose or throw away information**.

**REMARK**

Eigenvalues and eigenvectors definition:

An eigenvector of a square matrix $\mathbf{A}$ is a non-zero vector $v$ such that multiplying it by $\mathbf{A}$ alters only the scale of $v$:

$$\mathbf{A}v = \lambda v$$

The scalar $\lambda$ is the eigenvalue corresponding to this eigenvector. If $v$ is an eigenvector of $\mathbf{A}$, then so it is any rescaled vector $sv$ for $s \in \mathbb{R},  s \neq 0$. Moreover $sv$ still has the same eigenvalue. For this reason, we usually look for unit eigenvectors.

We can transform the original dataset so that the eigenvectors are the basis vectors and find the new coordinated of the data points with respect to this new basis

A set of vectors is termed a "basis" for a vector space if:

1. **Linear Independence**: No vector in the set can be represented as a linear combination of the other vectors.
2. **Spanning the Vector Space**: Every vector in the vector space can be represented as a linear combination of the basis vectors.

In the standard Cartesian plane (R^2), the basis vectors are often represented as:
- **i** = [1, 0]
- **j** = [0, 1]

For a three-dimensional space (R^3), the standard basis vectors are:
- **i** = [1, 0, 0]
- **j** = [0, 1, 0]
- **k** = [0, 0, 1]

However, these standard basis vectors are just one possible choice. Any set of vectors that satisfy the two properties above can be considered a basis for that vector space.

```python
# multivariate normal distribution
# mean
mu = [0,0]
# variance
# sigma = np.eye(2)                 # Standard bivariate normal distribution 
sigma = [[0.6,0.2],[0.2,0.2]]     # Correlated bivariate normal distribution
n = 1000
x = np.random.multivariate_normal(mu, sigma, n).T
len(x)
```

## Compute the principal component coordinates

Principal components are simply the eigenvectors of the covariance matrix used as basis vectors. Each of the original data points is expreessed as a linear combination of the principal components, giving rise to a new set of coordinates. Multiply eigenvectors and the original data to get the principal components.

PCA learns a linear projection that aligns the direction of greatest variance with the axes of the new space.

## Dimensionality reduction via PCA
To reduce the dimensionality of the problem, we simply set all but the first k principal component coordinates to zero. This is equivalent to ignoring all except the first k principal components.
In terms of the data it means projecting each vector $x_i$ in the k principal components subspace:
$$ x_i = \sum_{i=1}^d {pcc}_i v_i \sim  \sum_{i=1}^k {pcc}_i v_i.$$
In terms of the covariance matrix it means approximating $cov(X)$ as
$$cov(X) =  VDV^T = \sum_{i=1}^d e_i v_i\cdot v_i^T  \sim \sum_{i=1}^k e_i v_i\cdot v_i^T$$

We can represent the amount of variance in a dataset with just a couple of principal components.

Refer to this [Stack Exchange post](https://stats.stackexchange.com/questions/229092/how-to-reverse-pca-and-reconstruct-original-variables-from-several-principal-com) for useful additional explanation.

## PCA with sklearn

Principal component analysis (PCA) performs a linear transformation on the data so that most of the variance or information in your high-dimensional dataset is captured by the first few principal components. The first principal component will capture the most variance, followed by the second principal component, and so on.

The PCA method carries over from the field of statistics to unsupervised Machine Learning without any changes.

*PCA is used to decompose a multivariate dataset in a set of successive orthogonal components that explain a maximum amount of the variance.*


In scikit-learn, PCA is implemented as a transformer object that learns components in its fit method, and can be used on new data to project it on these components. However, we will explore the calculation around PCA on a simple toy problem, before applying it to a bigger dataset.
